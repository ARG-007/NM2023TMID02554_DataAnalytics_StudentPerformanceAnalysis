# House Price India Visualization

```python
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(150)
x = [x for x in range(len(ys))]

plt.scatter(x, ys, marker='o', c='b', label='Data Points')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Sample Visualization")
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('House Price India.csv')
df.head()
```

```python
df.info()
```

# **Univariate Analysis**

### Histogram


```python
plt.hist(df['Area of the house(excluding basement)'],bins=15)
plt.xlabel("Area of the house(excluding basement)")
plt.ylabel("Count")
```

### Displot


```python
sns.distplot(df['Area of the house(excluding basement)'],bins=5)
```

### Boxplot

```python
sns.boxplot(df['Price'])
```

### Violinplot

```python
sns.violinplot(x=df['lot_area_renov'])
```

# **Bivariate Analysis**

### Scatterplot

```python
sns.scatterplot(x=df['number of bedrooms'],y=df['lot_area_renov'])
```

### Joinplot

```python
sns.jointplot(data = df,x = 'Price',y = 'Number of schools nearby')
```

### Line plot

```python
plt.plot(df['Area of the basement'],df['Lattitude'])
```

# Multivariate Analysis

### Pairplot

```python
X = df[['number of bedrooms','number of bathrooms','lot area','living area','Price']]
sns.pairplot(X,dropna=True)
```

```python
df.drop(columns=['id','Date'],inplace=True)
sns.heatmap(df.corr(),annot=True)
```
